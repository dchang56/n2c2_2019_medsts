```python
import csv
import json
import os
import pandas as pd
```

```python
output_file = '/home/dc925/project/medsts/output/mtdnnbase_test/run_0/k_fold_4/errors.csv'
```

```python
df = pd.read_csv(output_file)
```

```python
dev_data_file = '/home/dc925/project/data/seq_pair/MEDSTS/k_fold_4/dev.jsonl'
```

```python
lines = []
with open(dev_data_file, 'r') as f:
    for line in f:
        lines.append(json.loads(line))
```

```python
sent1s = []
sent2s = []
for p in list(df.pid):
    for ex in lines:
        if int(ex['pid']) == p:
            sent1, sent2 = ex['sentence1'], ex['sentence2']
            sent1s.append(sent1)
            sent2s.append(sent2)
            
```

```python
df['sent1'] = pd.Series(sent1s)
df['sent2'] = pd.Series(sent2s)
```
